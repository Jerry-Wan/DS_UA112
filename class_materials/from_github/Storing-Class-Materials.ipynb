You can use the utility `tar` within the command line interface to package a collection of files into an archive. The archive contains name, content, time stamps, file access permissions, and directory organization of files.

#### Extracting

If you have `tar` on your computer, then you could use the following code to extract the files.

```python
!tar -xzvf 'ds_ua_112.tar.gz' -C 'DS-UA-112-JupyterHub/'
```

If you don't have `tar` on your computer, then you could use the following code to extract the files.

```python
import tarfile
import pathlib

pathlib.Path("DS-UA-112-JupyterHub/").mkdir(parents=True, exist_ok=True)

assert (pathlib.Path.cwd() / 'ds_ua_112.tar.gz').exists(), "Ensure that ds_ua_112.tar.gz is in your current working directory"

with tarfile.open('ds_ua_112.tar.gz', "r:gz") as dh:
    dh.extractall("DS-UA-112-JupyterHub/")
```
